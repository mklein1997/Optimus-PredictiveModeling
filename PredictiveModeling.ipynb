```python
# Import initial dependencies.

import pandas as pd
import warnings
warnings.simplefilter('ignore')
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in CSV.

voterfile = pd.read_csv("./voterfile.csv")
voterfile
```

```python
# Clean out unneccessary columns that do not directly reflect voting preference.

voterfile.drop(columns=['maritalstatus','dwellingtype','dma','occupationindustry','petowner_dog',
                        'intrst_nascar_in_hh','intrst_musical_instruments_in_hh','donates_to_liberal_causes',
                       'donates_to_conservative_causes','home_owner_or_renter'], inplace=True)
voterfile.head()
```

```python
# Create dummy variables for categorical variables to allow for model to read data

voterfile = pd.get_dummies(voterfile)
voterfile.columns
```

```python
party = pd.DataFrame(data=voterfile, columns=['party_Democratic', 'party_Republican'])
party.isna().any()
```

```python
# Check for NaN values and drop
cleansample = voterfile[['age', 'vh12p', 'vh12g', 'vh10g', 'vh08g','vh08p', 'cd', 'party_Democratic', 'party_Republican','vh14p']]
cleansample.dropna(inplace=True)
cleansample
```

```python
# Assign data and target variables

X = cleansample[['age', 'vh12p', 'vh12g', 'vh10g', 'vh08g','vh08p', 'cd', 'party_Democratic', 'party_Republican']]
y = cleansample[['vh14p']].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create scaled model to fit to training data

from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
# Transform data using scaled models

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# Create a linear regression model and fit it to scaled data

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
# Make predictions with scaled and fitted model

predictions = model.predict(X_test_scaled)
model.fit(X_train_scaled, y_train_scaled)
plt.scatter(model.predict(X_train_scaled), model.predict(X_train_scaled) - y_train_scaled, c="blue", label="Training Data")
plt.scatter(model.predict(X_test_scaled), model.predict(X_test_scaled) - y_test_scaled, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()
```

```python
# Calculate MSE and R-squared values
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test_scaled, predictions)
r2 = model.score(X_test_scaled, y_test_scaled)

print(f"MSE: {MSE}, R-squared: {r2}")

# First model yields pretty low accuracy when scaled. Ideally looking for accuracy close to 90%.
```

```python
# Next, build a model using KNN (Nearest Neighbors) method.

from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder
```

```python
# Train and test KNN model using previously outlined X and y parameters

# Separate labels and feature

X = cleansample.drop(labels=['vh12g'], axis=1)
y = cleansample['vh12g']

# sPlit data for train and test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Train model

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn.predict(X_test)

print(f"Test Accuracy: {knn.score(X_test, y_test)*100}%")

# Without scaling, this model already provides a much more accurate prediction. 
```

```python
# Use MinMaxScaler to scale down model to improve accuracy, similar method as linregress model.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
features = ['age', 'vh12p', 'vh12g', 'vh10g', 'vh08g','vh08p', 'cd', 'party_Democratic', 'party_Republican']

cleansample[features] = scaler.fit_transform(cleansample[features])
```

```python
# Run model again with scaled values

X = cleansample.drop(labels=['vh12g'], axis=1)
y = cleansample['vh12g']

# sPlit data for train and test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Train model

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn.predict(X_test)

print(f"Test Accuracy: {knn.score(X_test, y_test)*100}%")

# Scaled values have slightly improved accuracy, but nothing too significant, considering accuracy was already 
# fairly high originally. 
```

```python
# Taking a look at feature importance to determine which factors contribute most to model accuracy.

featureimportance = SelectKBest(score_func=chi2, k='all')
fit = featureimportance.fit(X, y)
fitscores = pd.DataFrame(fit.scores_)
fitcolumns = pd.DataFrame(X.columns)

# combine dataframes for readability

featurescoring = pd.concat([fitcolumns, fitscores], axis=1)
featurescoring.columns = ['Feature', 'Score']
print(featurescoring.nlargest(9, 'Score'))

# This feature importance chart shows that prior voting tendencies do the best job of predicting future voting habits.
```
